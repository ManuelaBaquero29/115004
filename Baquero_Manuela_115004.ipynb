```python
'''Este programa permite leer un número entero y escribirlo en reversa (Sin 
   utilizar comandos de ordenamiento). Por ejemplo, 461975 será 579164. '''

num = int(input('Escriba un número entero:'))
num_reducido = abs(num)      #Absoluto del número para facilidad de cálculos
dig_num = []                 #Lista que guarda los dígitos del número
aux = True                   #Variable auxiliar booleana para determinar cuando
                             # se detiene el ciclo while

while aux:
    dig = num_reducido%10    #Ciclo que permite guardar en dig_num los dígitos
    num_reducido //= 10      #del número
    dig_num.append(dig)
    if not num_reducido > 0:
        aux = False

#num_reversa guarda la conversión de cada elemento de dig_num a una cadena de 
#texto sin espacios       
num_reversa = "".join(str(x) for x in dig_num)

if num < 0:
    print(f'El número en reversa de {num} es {num_reversa}-')
else:
    print(f'El número en reversa de {num} es {num_reversa}')
```

```python
'''Este programa permite leer un número entero y determinar el número de cincos
   que están consecutivos en un número. Por ejemplo, en 5512355551 hay 6 cincos 
   consecutivos. Ahora, en 234555234512 solo hay 3 cincos consecutivos.'''

num = int(input('Escriba un número entero: '))                           
num_2 = abs(num) # Copia del valor absoluto de num para facilitar los cálculos

lista_num = []   #Lista que guarda los dígitos de num

while num_2 > 0:
    lista_num.append(num_2 %10)
    num_2 //=10

lista_num.append(0)  #Se agrega un 0 a la lista de dígitos para que no falle el
                     #el indexado
cantidad = 0         #Contador de 5 consecutivos
for i in range(len(lista_num)):
    if lista_num[i] == 5:
        if lista_num[i-1] == 5 or lista_num[i+1] == 5:
            cantidad +=1

print(f'Hay {cantidad} cincos consecutivos en el número {num}')
```

# Ejercicio 6

```python
'''Este programa permite leer un número entero y determinar si la suma de sus 
   dígitos es un número de Fibonacci.'''

while True:
    try:
        num=(int(input('Escriba un número entero:')))
        break
    except ValueError:
        print('Por favor escriba un número entero. Intente de nuevo...' )

if num < 0:
    print('Los números Fibonacci no pueden ser negativos')
else:
    #Variable auxiliar booleana que determina cuando se detiene el ciclo while
    aux = True           
    suma_digitos = 0  #Variable que guarda la suma de los dígitos de num
    while aux:        #Ciclo que permite encontrar la suma de lo dígitos de num
        dig = num%10
        num //= 10
        suma_digitos += dig
        if not num > 0:
            aux = False
            
    #Función que permite hallar los números fibonacci entre 0 y un número
    def fib(n):
        num_fibonacci = []
        a,b = 0,1
        while a<n:
            num_fibonacci.append(a)
            a,b = b, a+b
        return num_fibonacci   
    
    #Lista que guarda lo que retorna la función fib de suma_digitos, se le 
    #suma 1 para que la función tenga en cuenta el valor de suma_digitos
    lista_fibonacci = fib(suma_digitos+1)
    
    #Variable auxiliar booleana para determinar si el número es Fibonacci
    aux_2 = False    
    for i in lista_fibonacci:
        if suma_digitos == i:
            aux_2 = True
        else:
            aux_2 = False
    
    if aux_2 == True:
        print(f'{suma_digitos} es un número Fibonacci')
    else:
        print(f'{suma_digitos} no es un número Fibonacci')
```

# Ejercicio 7

```python
'''Este programa permite leer un número entero positivo y determine si existen 
   en el número dígitos repetidos y reporta cuales son los dígitos 
   repetidos.'''

while True:
    try:
        num=int(input('Introduzca un número entero positivo: '))
        if num>0:
            break
        else:
            raise ValueError
    except ValueError:
        print('Por favor ingrese un número entero, intente de nuevo.')

digitos = [] #Lista que guarda los dígitos de num
aux = True   #Variable booleana que determina cuando se detiene el ciclo while

#Se determina cada dígito del número ingresado
while aux:
    dig = num%10
    num //= 10
    digitos.append(dig)
    if not num > 0:
        aux = False

leidos = []     #Lista que guarda el dígito i leído en el primer for   
repetidos = []  #Lista que guarda el dígito repetido y cuantas se repite
   
for i in digitos:
    cont=0      # Repeticiones del dígito i
    if i not in leidos:
        leidos.append(i)
        for j in digitos:
            if i == j:
                cont +=1
    if cont >=2:
        repetidos.extend([[i,cont]])
        cont = 0

if len(repetidos) == 0:
    print(f'No hay números repetidos en el número ingresado')
elif len(repetidos) == 1:
    print(f'En el número ingresado se repite {repetidos[0][1]} veces el dígito\
 {repetidos[0][0]}')
else:
    print('En el número ingresado se repiten los siguientes números:\n')
    for i in repetidos:
        print (f'{i[0]} se repite {i[1]} veces')
```

# Ejercicio 11

```python
'''Este programa recibe una cadena de texto y reporta:
   a) cuántas letras vocales en mayúscula se entraron;
   b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
   c) cuántos dígitos se entraron,
   d) cuántos espacios se entraron;
   e) cuántas palabras reservadas se entraron.
   Utilizar para ello los métodos de cadenas especializados como .isupper(), 
   .isdigit(), etc. y la palabra reservada in.'''

# Se importa la librería que permite leer todas las palabras reservadas.
import keyword 

# Se pide el texto a ingresar.
texto = input('Digite una cadena de texto: ')

# Se almacenan los posibles caracteres.
vocales = 'AEIOUÜ'
tildes = 'áéíóúÁÉÍÓÚ'

# Variables que almacenan las cantidades.
num_vocales_may = 0
num_tildes = 0
num_digitos = 0
num_espacios = 0
num_reservadas = 0      

# Se determina la cantidad de letras, dígitos o espacios.
for i in texto:
    if i in vocales:
        num_vocales_may += 1
    elif i in tildes:
        num_tildes += 1
    elif i.isdigit():
        num_digitos += 1
    elif i == ' ':
        num_espacios += 1

#Determinación de la cantidad de palabras reservadas en el texto.
# S.split() divide la cadena en palabras en base a los espacios.
#Keyword.kwlist es una lista de las palabras reservadas de Python.
texto_list = texto.split(' ')
for i in range(len(texto_list)):
    if texto_list[i] in keyword.kwlist:
        # Separa la cadena entre espacios para conformar una palabra.
        num_reservadas +=1
    
# Impresión de las cantidades.
print(f'en el texto ingresado hay: ')
print(f'{num_vocales_may} vocales en mayúscula')
print(f'{num_tildes} letras con tilde')
print(f'{num_digitos} dígitos')
print(f'{num_espacios} espacios')
print(f'{num_reservadas} palabras reservadas')
```

```python
'''Este programa permite leer dos palabras (podrían estar en mayúscula o
   minúscula) y determinar cual está primero en el diccionario. El programa 
   soporta letras con tilde, la ü y la ñ.
   Orden alfabético: aAáÁ'''
   
# Lectura de las dos palabras y se almacenan en listas
palabra1 = list(input('Digite la primera palabra:'))
palabra2 = list(input('Digite la segunda palabra:'))

#Se pasa la lista a una cadena para poder imprimir la palabra normalmente
palabra1_cad = "".join(str(x) for x in palabra1)
palabra2_cad = "".join(str(x) for x in palabra2)

#Lista que tiene las letras en orden del alfabeto
alfabeto = ['a','A','á','Á','b','B','c','C','d','D','e','E','é','É','f','F',\
            'g','G','h','H','i','I','í','Í','j','J','k','K','l','L','m','M',\
            'n','N','ñ','Ñ','o','O','p','P','q','Q','r','R','s','S','t','T',\
            'u','U','ü','Ü','v','V','w','W','x','X','y','Y','z','Z']

palabra_1_num = [] #Lista que guardará el indexado cada letra de la palabra1
for i in palabra1:
    indexado1 = 0  #indexado de cada letra
    for j in alfabeto:
        if i == j:
            indexado1 = alfabeto.index(j)
            palabra_1_num.append(indexado1)
            
palabra_2_num = [] #Lista que guardará el indexado cada letra de la palabra1
for i in palabra2:
    indexado2 = 0  #indexado de cada letra 
    for j in alfabeto:
        if i == j:
            indexado2 = alfabeto.index(j)
            palabra_2_num.append(indexado2)


if palabra_1_num < palabra_2_num:
    print(f'"{palabra1_cad}" está primero en el diccionario que \
"{palabra2_cad}"')
elif palabra_1_num > palabra_2_num:
    print(f'"{palabra2_cad}" está primero en el diccionario que \
"{palabra1_cad}"')
elif palabra_1_num == palabra_2_num:
    print(f'"{palabra1_cad}" y "{palabra2_cad}" son la misma palabra, están en\
 la misma posición en el diccionario')
```

```python
'''Escribir un programa que lea una cadena de texto y la imprima como un 
   triángulo. Por ejemplo, si la cadena entrada es: "Es una locura odiar a 
   todas las rosas porque una te pinchó: renunciar a todos tus sueños porque 
   uno de ellos no se realizó - El Principito", el programa debe imprimir:
            E
           sun
          alocu
         raodiar
        atodaslas
       rosasporque
      unatepinchó:r
     enunciaratodost
    ussueñosporqueuno
   deellosnoserealizó-
      ElPrincipito
'''
# Se lee la cadena
cadena = input('Digite la cadena de texto: ')

# Se eliminan los espacios presentes en la cadena
cadena = cadena.replace(' ', '')

# Se determina el número de caracteres en la cadena
num_caract = len(cadena)

# Se determina el número de filas y el número máximo de caracteres por fila
temp = 0                       # Variable auxiliar 
num_filas = 1                  
while temp < num_caract:
    temp += 2 * num_filas - 1
    num_filas += 1

num_max = 2 * num_filas - 1    # Número máximo de caracteres en una fila

# Se imprime la cadena de texto 
aux_1 = 0        #Almacena la última posición en la que se imprime la fila i
aux_2 = 0        #Cantidad de letras que se imprimen en la fila i
contador = 0     # Verifica si luego de imprimir el triángulo aún quedan letras
                 # por imprimir

for i in range(1,num_filas):
    aux_2 += 2*i-1     
    print(cadena[aux_1:aux_2].center(num_max,' '))
    aux_1 = aux_2      
    contador += 1
    
if contador != num_caract:
    print(cadena[aux_2:].center(num_max,' '))
```

```python
'''Este programa permite leer una lista de números enteros y enumerar cuantos 
   elementos se repiten exactamente dos veces. Utilizar para tal fin el método
   count() de listas.'''
   
print('Por favor ingrese la lista dentro de [] y separe los elementos con \
comas. Por ejemplo:[1,2,3]',end = '')
lista = eval(input('Ingrese una lista de números enteros:'))

repetidos = [] # Lista que guarda los números que se repiten 2 veces

for i in lista:
    if lista.count(i) == 2:
        repetidos.append(i)

for i in repetidos:
    if lista.count(i) > 1:
        repetidos.remove(i)
        
if len(repetidos) == 0:
    print(f'No hay números repetidos en la lista ingresada {lista}')
elif len(repetidos) == 1:
    print(f'El {repetidos[0]} está repetido 2 veces')
else:   
    print(f'Los números {repetidos} están repetido 2 veces')
```

```python
'''Este programa permite leer una lista y determinar cuál es el segundo 
   elemento más grande de dicha lista. No utilice ningún tipo de algoritmo de 
   ordenamiento para realizar este ejercicio.'''
   
print('Por favor ingrese la lista dentro de [] y separe los elementos con \
comas. Por ejemplo:[1,2,3]')
print("Si quiere ingresar una letra o una palabra escribala entre comillas \
simples ''. Por ejemplo: 'a' 'Mañana'")
lista = eval(input('Ingrese una lista:'))

#Esta función permite hallar el elemento más grande de una lista
def mayor(lista):
    mayor = lista[0]
    for i in lista:
        if mayor < i:
            mayor = i
    return mayor

lista_1 = lista.copy() #Copia de lista

if len(lista) == 0:
    print('Ingresó una lista vacía, por lo tanto no hay segundo elemento')
elif len(lista) ==1:
    print(f'La lista solo tiene 1 elemento, por lo tanto no hay segundo \
elemento')
else:
    mayor_list = mayor(lista_1) #Elemento más grande de lista lista
    lista_1.remove(mayor_list)
    segundo = mayor(lista_1)    #Segundo elemento más grande de lista
    print(f'El segundo elemento más grande de {lista} es {segundo}')
```
